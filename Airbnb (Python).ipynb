```python
# so I can drop bathrooms and calender_upadted cols
# all values similar; no use
df = listings.drop(['bathrooms', 'calendar_updated', 'has_availability'], axis=1)
```

```python
#Some other columns containing ids and urls are also not needed
df = df.drop(['scrape_id', 'host_id', 'listing_url', 'last_scraped', 'host_url',
              'host_thumbnail_url', 'host_picture_url', 'picture_url'], axis=1)
```

```python
#I will only keep the 2 main columns of 'minimum_nights' and 'maximum_nights' and discard other variations.
cols_to_drop = ['minimum_minimum_nights', 'maximum_minimum_nights',
               'minimum_maximum_nights', 'maximum_maximum_nights', 'minimum_nights_avg_ntm',
               'maximum_nights_avg_ntm']
df.drop(cols_to_drop, axis=1, inplace=True)
```

```python
#We will start by processing the categorical columns.
#host_since: Datetime column telling us when the host joined the platform
#derive a new column conveying the number of days since the host is active on the platform. We subtract the values in the host_since column with the date when the data was scraped. In our case, this is 9/10/2022.

from datetime import datetime
df.host_since = pd.to_datetime(df.host_since)
df['Active_Days'] = (datetime(2022,9,10) - df.host_since).astype('timedelta64[D]')
df.Active_Days.fillna(df.Active_Days.median(), inplace = True)
```

```python
#extract no of bathrooms from text
import sys
import re

def get_bath_counts(column):
    try:
        if 'half' in column or 'Half' in column:
            a = 0.5
        else:
            a = re.findall(r'\d+', column)[0]
    except:
        print(column)
        prin(sys.exc_info()[0])
    return a

df['Num_of_Bathrooms'] = pd.to_numeric(df['bathrooms_text'].apply(lambda x: get_bath_counts(x)))
```

```python
#property_type: describes the kind of property.
df.property_type.value_counts()
```

```python
#bin the property types into 4 categories: private_room, shared_room, entire_space and other.
def bin_property_type(column):
    if 'Private' in column:
        valu = 'Private Room'
    elif 'Shared' in column:
        valu = 'Shared Room'
    elif 'Entire' in column:
        valu = 'Entire Space'
    elif 'hotel' in column or 'aparthotel' in column:
        valu = 'Hotel'
    elif 'Private' not in column or 'Shared' not in column or 'Entire' not in column or 'hotel' not in column or 'aparthotel' not in column:
        valu = 'Other'
    return valu

df['property_type'] = df.property_type.apply(lambda x : bin_property_type(x))
```

```python
## fill the nan values in these columns with the median value of the column
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
df['beds'] = df['beds'].fillna(df['beds'].median())
df['reviews_per_month'] = df['reviews_per_month'].fillna(df['reviews_per_month'].median())
```

```python
#column 'host_total_listings_count' that is similar to 'host_listings_count', so this can be dropped.
df.drop(['host_total_listings_count'], axis = 1, inplace=True)
df = df.dropna(subset=['host_listings_count'])
```
